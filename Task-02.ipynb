```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting only numeric columns from the dataset
numeric_columns = df.select_dtypes(include='number')

# Creating a boxplot for numeric columns
sns.boxplot(data=numeric_columns, palette='magma')
plt.title('Boxplot for Numerical Columns')  # Corrected
plt.show()
```

```python
# Selecting specific columns
columns = ['Age', 'Fare']

# Calculating quartiles and IQR for each column
for col in columns:
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Filtering outliers for each column
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Creating a boxplot after removing outliers
sns.boxplot(data=df, palette='magma')
plt.title('Boxplot for dataset after removing outliers')
plt.show()
```
